По сути два указателя. Проходм подмассивом фиксированного (иногда динамического) размера и че то считаем.

Два типа:
1. Динамический
2. Фиксированный

Фиксированный:

```python
def max_sum_subarray(arr, k):
    n = len(arr)
    
    if n < k:
        return 0
    
    window_sum = sum(arr[:k])
    max_sum = window_sum
    
    
    for i in range(k, n):
        window_sum += arr[i] - arr[i - k]
        max_sum = max(max_sum, window_sum)
        
    return max_sum
```

```python
arr = [2, 1, 5, 1, 3, 2]
k = 3

max_sum = max_sum_subarray(arr, k)
max_sum
```

## __Практика__

```python
def maxProfit(prices):
    l = 0
    r = 1
    
    max_prof = 0
    
    while r < len(prices):
        if prices[r] > prices[l]:
            max_prof = max(max_prof, prices[r] - prices[l]) 
        else:
            l = r
            
        r += 1
        
    return max_prof
```

```python
maxProfit([1,2,4,2,5,7,2,4,9,0,9])
```

```python
def minSubArrayLen(target, nums):
    l = 0
    r = 0
    win_len = 1
    min_len = 1e9
    win_sum = nums[0]
    
    if win_sum >= target:
        return 1
    
    while r < len(nums)-1:
        if win_sum >= target:
            min_len = min(min_len, win_len)
            win_len -= 1
            win_sum -= nums[l]
            l += 1   
        else:
            r += 1
            win_sum += nums[r]
            win_len += 1
    else:
        if l < r:
            while l <= r:
                if win_sum >= target:
                    min_len = min(min_len, win_len)
                win_len -= 1
                win_sum -= nums[l]
                l += 1   
                
    if min_len == 1e9: return 0
    
    return min_len
```

```python
minSubArrayLen(7, [1,1,1,1, 7])
```
