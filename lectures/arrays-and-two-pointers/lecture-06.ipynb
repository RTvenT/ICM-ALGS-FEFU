# __Хэш-таблицы__

```python
class Solution:
    def wordPattern(self, pattern: str, s: str) -> bool:
        d1 = dict()
        d2 = dict()

        l = s.split()
        
        if len(l) != len(pattern): 
            return False

        for i in range(len(l)):
            pat = pattern[i]
            word = l[i]
            
            if not d1.get(pat) and d2.get(word):
                return False
    
            if not d1.get(pat) and not d2.get(word):
                d1[pat] = word
                d2[word] = pat
            elif d1[pat] != word or pat != d2[word]:
                return False
            
            
        return True
```

```python
solve = Solution()

solve.wordPattern("abba", "dog dog dog dog")
```

```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        d1 = dict()
        d2 = dict()
        
        if len(s) != len(t): 
            return False

        for i in range(len(s)):
            pat = s[i]
            word = t[i]
            
            if not d1.get(pat) and d2.get(word):
                return False
    
            if not d1.get(pat) and not d2.get(word):
                d1[pat] = word
                d2[word] = pat
            elif d1[pat] != word or pat != d2[word]:
                return False
            
            
        return True
```

```python
solve = Solution()

solve.isIsomorphic("paper", "title")
```
