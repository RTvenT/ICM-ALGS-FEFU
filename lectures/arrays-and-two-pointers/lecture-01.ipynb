## __"Нотация O-Большое, Git/GitHub"__

### __O-Большое__

1. O(1) - Константная сложность, время выполнение не зависит от размера входных данных (моментально)
2. O(logn) - Логарифмическая чуть медленнее
3. O(n) - Линейная, напрямую от размера входных
4. O(nlogn) - Квазилинейная (TimSort, HeapSort)
5. O(n^2) - ГОВНО
6. O(2^n) - ГОВНО ГОВНО
7. O(n!) - ГОВНО ГОВНО ГОВНО

- Константы отбрасываются: O(3*n^2 + 5) ~ O(n^2)
- Меньшее слагаемое тоже: O(n^2 + logn + 4) ~ O(n^2)

```python
def intersection(arr1: list, arr2: list) -> list: 
    return list(set(arr1) & set(arr2))
```

```python
arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))

print(arr1, arr2, sep='\n')
print(intersection(arr1, arr2))
```
