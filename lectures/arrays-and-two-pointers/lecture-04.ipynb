Valid Parentheses (#20 LeetCode)

```python
def isValid(s: str) -> bool:
        stack = []
        d = {
            '(': ')',
            '{': '}',
            '[': ']'
        }
    
        for p1 in s:
            if p1 in d.keys(): 
                stack.append(p1)
            elif stack:
                p2 = stack.pop()
                if p1 != d[p2]:
                    return False
            else:
                return False
                
        return not bool(len(stack))
```

```python
def removeStars(s: str) -> str:
        stack = []

        for c in s:
            if c  != '*':
                stack.append(c)
            else:
                stack.pop()

        return ''.join(stack)
```
