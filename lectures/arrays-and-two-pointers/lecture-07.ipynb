## __Bin Search__

```python
a = [1, 2,34 ,4,32 ,2, 4,3,5 ,32,45,4, 2,4, 2,42,3, 4,34, 3]

value = int(input())

# индексы первого элемента, последнего и среднего
low = 0
high = len(a) - 1
mid = len(a) // 2

while a[mid] != value and low <= high:
    if value > a[mid]:
        low = mid + 1
    else:
        high = mid - 1
    mid = (low + high) // 2

if low > high:
    print('No value')
else:
    print('ID =', mid)
```

```python
def searchMatrix(matrix, target) -> bool:
    m = len(matrix)
    n = len(matrix[0])
    
    low = 0
    high = m * n - 1
    
    while low <= high:
        mid = (low + high) // 2
        
        row = mid // n  
        col = mid % n   
        
        if target > matrix[row][col]:
            low = mid + 1
        elif target == matrix[row][col]:
            return True
        else:
            high = mid - 1
    
    return False
```

```python
matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 3

searchMatrix(matrix, target)
```
