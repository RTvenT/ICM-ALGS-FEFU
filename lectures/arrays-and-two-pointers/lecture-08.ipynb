# __Связанные списки__

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def deleteDuplicates(self, head):
        cur = head.next
        prev = head
        
        while cur:
            if cur.val == prev.val:
                prev.next = cur.next

            cur = prev.next
            
        return head
```
